## Time Resampling

```python
df_tesla.head(2)
```

### Here the below code gives in each year what was the min value

```python
df_tesla.resample(rule='A').min()
```

## Simplilarly over here it gives max in each year

```python
df_tesla.resample(rule='A').max()
```

### We have ploted a graph of max in each year wrt Open attribute

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

## Same thing the plot is wrt Volume

```python
df_tesla.resample(rule='A').max()['Volume'].plot()
```

## So let's see what does this rule='A' signify year end frequency ie you are going to get year end data

## Suppose you want quaterly start frequency

```python
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

### Business end frequency

```python
df_tesla.resample(rule='BA').max()['High'].plot()
```

### Business quaters

```python
df_tesla.resample(rule='BQS').max()['High'].plot()
```

```python
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

### Rule M is wrt Monthy data

```python
df_tesla['Open'].resample(rule='M').mean()
```

```python
df_tesla['Open'].resample(rule='M').max()
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

##### The resample() method in pandas is used to change the frequency of a time series. It's particularly useful when you have time series data with irregular intervals and you want to convert it into a series with a fixed frequency (e.g., converting daily data to monthly data).

### Let's understand 2 main keywords rolling & explanding

```python
df_tesla['High'].rolling(10).mean().head(20)
```

### So in the above code the rolling is applied for the feature High in df what it does is it considers the number -1 mentioned rows and find the aggregate over here it is mean so it finds mean and then go next and the same thing continues you can observe it next code block

```python
df_tesla['High'].rolling(11).mean().head(20)
```

### ||y you can apply for max,min...

```python
df_tesla['High'].rolling(10).max().head(20)
```

### Let's add a column open 30 days rolling into our df

```python
df_tesla['Open 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(31)
```

### So now let's plot a graph between Open and Open 30 days rolling

```python
df_tesla[['Open','Open 30 days rolling']].plot(figsize=(12,5))
```
